**ÖRÜNTÜ TANIMA-SINIFLANDIRMA-LOJİSTİK REGRESYON**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
dataset = pd.read_csv("heart_failure_clinical_records_dataset.csv")
df = dataset.copy()
df
```

```python
df["DEATH_EVENT"].value_counts()

#203 tane 0 değeri
#96 tane 1 değeri
```

```python
#istatistik veri
df["DEATH_EVENT"].value_counts().plot.barh();
```

```python
df.describe().T
```

```python
#X = dataset.iloc[:,:-1].values #tüm satır ve sütunları al, son kolon hariç
#y = dataset.iloc[:,-1].values #tüm satırlar ve son kolon al
y=df["DEATH_EVENT"]
X=df.drop(["DEATH_EVENT"], axis=1)
```

```python
#stats model
loj = sm.Logit(y,X)
loj_model = loj.fit()
loj_model.summary()
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
#veriyi test ve train olarak ayırma
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=60)
```

```python
X_train
```

```python
#ÖLÇEKLENDİRME
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
print(X_train) #train için - her biri için standartize edilmiş değerler
```

```python
X_test = sc_X.transform(X_test)
print(X_test) # test için - her biri için standartize edilmiş değerler
```

```python
#MODELLEME --> LOJİSTİK REGRESYON
#ismi lojistik regresyon ama sınıflandırma 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(X_train, y_train) #fit ile model oluşturduk
```

```python
#TAHMİN İLE SONUÇ BULMA AŞAMASI
#y_pred --> tahmin verileri
y_pred = model.predict(X_test)
#predict --> hangi veri üzerinde predict X_test üzerinde 
y_pred
```

```python
#test verileri
print(y_test)
```

```python
#conf matrix
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
#confusion_matrix(y_true = y_test, y_pred = y_pred)
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
#%78 accuracy
```
