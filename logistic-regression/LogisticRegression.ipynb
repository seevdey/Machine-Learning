```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#veri setinde ID,cinsiyet, maas, yas var bunlara göre kredi verip verilmeyeceğini buluyoruz
#kesikli veri söz konusu --> kredi verecek miyim vermeyecek miyim
# y değişkenleriniz kesikli veri (kredi sütunu) --> ya 0 ya 1 
#ismi lojistik regresyon ama sınıflandırma yapıyoruz --> en temel şey bu
#kesinlikle sürekli bi veri değil

#fonksiyonun negatif olmaması için mutlak değeri alabiliriz, üstel fonk olabilir
#x değeri değişiyo --> maaş,yaş, cinsiyet
```

```python
veriseti = pd.read_csv("kredi_veriseti.csv")
df = veriseti.copy()
df.head(20) #20 satır yazdır
```

```python
#ben bi tane fonksiyon yapayım ki bu fonksiyona bu değerleri verdiğim zaman (maaş,yaş, cinsiyet)
#bana bi olasılıkla bi değer olsun. 
#Threshold 0 dan büyükse kredi ver, küçükse verme
#1/1+e-x
```

```python
def logistic(x):
    return 1.0/(1+np.exp(-x)) #buraya X değerlerini vericem logistik fonk bana 0 le 1 arasında değerler verecek

x = np.linspace(-10, 10, 1000)#-10 ile 10 arasında 1000 tane değer üret x e koy
y = logistic(x)

y #0 ile 1 arasında değerler
```

```python
plt.plot(x,y) #grafik bütün değerler 0 ile 1 arasında değer verir
#basamak fonksiyonu da denir
```

```python
df.drop('User ID', axis=1, inplace=True) #User ID işimize yaramaz kaldırdık
#inplace=true --> sadece drop edip belli bi değere attığınız zaman kullanabilirsiniz, kalıcı olmasını istersen True demen lazım
df.head()
```

```python
#ÖNİŞLEME, ENCODİNG
```

```python
df['cinsiyet'].values.reshape(-1,1) #numpy arrayini sütun vektörü olarak ver
#df['cinsiyet']
```

```python
#encoding işlemi çıktı veri setinde
#cinsiyet kısmını 0,1 yapmamız lazım
from sklearn.preprocessing import OneHotEncoder
oneHot = OneHotEncoder(handle_unknown='ignore')
oheCinsiyet = oneHot.fit_transform(df['cinsiyet'].values.reshape(-1,1)).toarray()
oheCinsiyet
```

```python
oneHot.categories_ #1 erkek, 0 kadın oldu
#işlem yapmak için numpy kullandık
```

```python
#sonra pandas a çevirdik
ohe_df = pd.DataFrame(oheCinsiyet, columns=oneHot.categories_)
ohe_df.head()
```

```python
ohe_df2 = df.join(ohe_df) #ohe_df ile df i birleştircez
ohe_df2
```

```python
#cinsiyeti silelim
ohe_df2.drop('cinsiyet', axis=1, inplace=True)
#axis=1 satır değil kolonu kaldır
```

```python
ohe_df2.head()
```

```python
X = ohe_df2.drop("kredi", axis=1)
y = ohe_df2['kredi']
```

```python
X
```

```python
y
```

```python
X.shape #X 400 satır, 4 kolondan luşuyo
```

```python
#test ve train datası olarak ayırma
from sklearn.model_selection import train_test_split
#veriyi test ve train olarak ayırma
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=60)
#random_state vermezsen her metodu çalıştırdığımda farklı sonuçlar çıkar
#%33 test, %67 train için ayrıldı
```

```python
X_train #%33 test, %67 train için ayrıldı
```

```python
X_train.shape #268 satır, 4 sütun var
```

```python
X_test.shape #132 satır, 4 sütun var
```

```python
y_test.shape #132 satır, 1 sütun var 
#y sonuç değeri, kredi verip verilmeyeceği
```

```python
#ÖLÇEKLENDİRME
from sklearn.preprocessing import MinMaxScaler, StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train) #yaş ve maaşı 0 ve 1 arasına çekecek
X_test = sc.fit_transform(X_test)
```

```python
print(X_test)
```

```python
#MODELLEME --> LOJİSTİK REGRESYON
#ismi lojistik regresyon ama sınıflandırma 
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=60)
model.fit(X_train, y_train) #fit ile model oluşturduk
```

```python
#TAHMİN İLE SONUÇ BULMA AŞAMASI
#y_pred --> tahmin verileri
y_pred = model.predict(X_test)
#predict --> hangi veri üzerinde predict X_test üzerinde 
y_pred
```

```python
#test verileri
y_test.values
```

#PERFORMANS KRİTERLERİ

```python
#conf matrix
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
#confusion_matrix(y_true = y_test, y_pred = y_pred)
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
#%81 doğruluk değeri çıktı
```

```python
accuracy_score(y_test, y_pred)
#%81 accuracy
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model, X_test, y_test, cv=10).mean()
#crass val ile farklı farklı 10 tane değer buldu, mean 
```
